```python
# Alcuni pacchetti che ci serviranno:
import numpy as np
# per gestire array di numeri
import matplotlib.pyplot as plt
# per fare grafici
import ipywidgets as ipy 
# per rendere i grafici interattivi
```

```python
# Creiamo una lista di 200 numeri equispaziati 
# fra 0 e 5:
x = np.linspace(0, 5, num=200)

# Scegliamo la nostra lunghezza d'onda
lam = 2
```

```python
# Facciamo il grafico dell'equazione d'onda
# senza la dipendenza temporale

y = np.sin(2 * np.pi * x / lam)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid("on")
```

```python
# Inseriamo ora la dipendenza dal tempo
# definiamo il periodo temporale T
T = 0.5

def plot(t):
    y = np.sin(2 * np.pi *(x/lam - t/T))
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid("on")

# Creiamo un grafico interattivo 
# con uno slider per il parametro t

ipy.interact(plot, t=(0, 2, 0.01))

# L'onda si muove verso destra quando il tempo aumenta!
```
